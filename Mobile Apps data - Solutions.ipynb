```python
# Check for duplicate records
dupe = apps_data.duplicated().sum()
print('In the data set having duplicated records of \n',dupe)
```

```python
# We need to drop all the duplicates 
print('Before dropping duplicates the shape of dataset is: ',apps_data.shape,'\n')

# Now drop the duplicates
apps_data.drop_duplicates(inplace =True)

print('After dropped all the duplicated records now the shape is: ',apps_data.shape)
```

### Inferences:

```python
# Check the unique categories of column 'Category'
print('Number of unique categories:',apps_data['Category'].nunique(),'\n')
print('The unique categories in that column:',apps_data['Category'].unique())
```

```python
# There is only one category is named 1.9 that should not be a category
# We need to remove that categor and their records
apps_data['Category'].replace('1.9',np.nan,inplace=True)
```

```python
# Check there is missing value  
print('Before dropping that invalid category :',apps_data['Category'].isna().sum(),'\n')

# Now that missing value should be drop
print('After dropping that invalid category :',apps_data['Category'].dropna(inplace = True),'\n')

# And finally check the number of unique categories
print('After dropping that invalid category no of categories is:',apps_data['Category'].nunique())
```

### Inferences:
* In the column 'Category' having 34 number of categories.
* And there is only one category is the numerical that should not be a category.
* So we need to drop that invalid category.
* I had replace that one with null value and then drop tha null value from the column.

```python
# Check the the missing values in the column 'Rating' and drop them 
print('Before treat the missing values:',apps_data['Rating'].isna().sum())
```

```python
# There are 1465 number of missing values its above 10% of the values are missing
# We cannot drop all those values so, we can impute with median
apps_data.Rating.fillna(apps_data.Rating.median(),inplace = True)
```

```python
# After treating all the null values
print('After treating the missing values',apps_data['Rating'].isna().sum())
```

```python
# Create a new column Rating_category
vals = apps_data.Rating.values
apps_data['Rating_category'] = np.where(apps_data['Rating']>3.5 , "High", 'Low')

# Check the added column
apps_data.head(16)
```

### Inferences:
* In the column 'Rating'I cheked for missing values there are 1465 is found.
* It cannot be dropped because of the missing values percentage in the dataset is more.
* So, we can impute them with median.
* Then the new column 'Rating_category'is created based on the ratings.
* The ratings is greater than 3.5 that labeled 'High' and lesser than is labeled 'Low'.

```python
# Convert the column 'Reviews' to numeric
# There is some object is present 

temp = pd.DataFrame(apps_data['Reviews'].str.isdigit())
temp[temp['Reviews'] == False]
```

```python
# The object is
apps_data.Reviews[10472]
# We need to replace as a numeric value
apps_data.Reviews.replace('3.0M',3.0 ,inplace = True)
```

```python
# NOw convert the column to numeric
apps_data['Reviews'] = apps_data['Reviews'].astype('int64')
```

```python
# Find the outliers by box plot
rev = apps_data['Reviews']
sns.boxplot(rev)
plt.xlabel('Reviews')
plt.ylabel('Counts')
plt.title('Box plot for Reviews')
plt.show()
```

```python
# from the box plot there are so many outliers present in the data 

# We need to standardised the data by using log tranformation
from sklearn.preprocessing import FunctionTransformer
log = FunctionTransformer(np.log1p)

# We need to create a column by logtransformer
apps_data['Reviews_std'] = log.fit_transform(apps_data['Reviews'])
apps_data['Reviews_std'].head()
# Now it converted in to standard scale
```

```python
# Now we create a box plot which does not contains outliers
sns.boxplot(apps_data['Reviews_std'])
plt.xlabel('Reviews')
plt.ylabel('Counts')
plt.title('Box plot for Reviews')
plt.show()
```

### Inferences:
* The above box plot is created by the column 'Reviews_std' which is derived from reviews column.
* Now there is no outliers

```python
# We need to find something from the column 'Size'

# First we should easily change the str to null values
apps_data['Size'] = apps_data['Size'].replace('Varies with devices',np.nan)

print('Before',apps_data.isna().sum().sum())
apps_data.dropna(inplace = True)
print('After',apps_data.isna().sum().sum())
```

```python
apps_data.shape
```

```python
# Check the column installs

# It contains some symbols like +

# Here i define function to extract the numeric values from the column
def num(elements):
    numb = ''.join(filter(str.isnumeric,elements))
    return numb
apps_data['Installs'] = [num(x) for x in apps_data['Installs']]

# Change the datatype of the column
apps_data['Installs'] = apps_data['Installs'].astype('float64')

apps_data['Installs'].head()
```

### Inferences:
* From the column installs the unwanted symbols are removed.
* And that column datatype is also changed.

```python
# Check the column Price
print('Number of Unique values:',apps_data['Price'].nunique())

# There are 92 types of various price values, now check all the values are numeric
def nums(elem):
    numbs = ''.join(filter(str.isnumeric,elem))
    return numbs
apps_data['Price'] = [nums(x) for x in apps_data['Price']]

# We need to change the datatype

apps_data['Price'].unique()
```

```python
# Now change the data type
apps_data['Price'] = apps_data['Price'].astype('int64')
apps_data['Price'].info()
```

### Inferences:
* The numerical values are extractded from the column.
* And all the numeric values are printed in the column 'Price'.
